```python
import random
import torch
import torch.nn as nn
import torch.optim as optim


import random

def generate_random_names( num_names):
    random.seed(42)  # Set a seed for reproducibility
    random_names = []
    for _ in range(num_names):
        first_name = random.choice(first_names)
        last_name = random.choice(last_names)
        random_names.append(f"{first_name} {last_name}")
    return random_names

first_names = ["Ali", "Zahra", "Reza", "Sara", "Mohammad", "Fatemeh", "Hossein", "Maryam", "noshin", "Narges", "Hamed", "Roya", "vahid", "majid", "marjan", "farzaneh", "sepideh", "maedeh", "nahid", "rosha"]
last_names = ["Ahmadi", "Hosseini", "Karimi", "Rahimi", "Hashemi", "Ebrahimi", "Moradi", "rohani", "Rostami", "Fazeli", "Hosseinzadeh", "Niknam", "armani", "karbasi", "pahlevan", "sadegh", "azari", "asghari", "salem", "davami"]

num_names = 400
random_names = generate_random_names( num_names)

# Display the generated names
for i in range(0, len(random_names), 12):
    print(", ".join(random_names[i:i+12]))
```

```python
import random
import torch

# Function to generate a conflict matrix between people
def Matrix_conflict_people(num_names:int, m_conflict : int, group_numbers):
    conflict_people = torch.zeros(num_names, num_names)
    count = 0
    incompatible_pairs = []
    while count < 600:
        i = random.randint(0, 399)
        j = random.randint(0, 399)
        if i in group_numbers != j in group_numbers and conflict_people[i][j] == 0:
            conflict_people[i][j] = 1
            count += 1
            # Adding the incompatible pairs with new format
            incompatible_pairs.append((i, j))

    # Formatting the incompatible pairs as required
    formatted_pairs = [f"( i :{pair[0]} , j :{pair[1]})" for pair in incompatible_pairs]
    return conflict_people.float(), formatted_pairs

# Setting print options for PyTorch
torch.set_printoptions(precision=0, linewidth=200)

# Example usage of the function
group_numbers = [i for i in range(400)]  # Assuming all 400 numbers are in group_numbers
conflict_matrix, incompatible_pairs = Matrix_conflict_people(400, 600, group_numbers)

# Displaying the result
print(conflict_matrix)
for pair in incompatible_pairs:
    print(pair)

```

```python
def group_conflict_matrix(groups_number, matrix_conflicts):

    num_groups = len(groups_number)
    group_conflicts = torch.zeros(num_groups, num_groups)
    
    xs, ys = [], []

    for i, group1 in enumerate(groups_number):
        for j, group2 in enumerate(groups_number):
            if i != j:
                conflicts = 0
                for member1 in group1:
                    for member2 in group2:
                        conflicts += matrix_conflicts[member1, member2]
                group_conflicts[i, j] = conflicts
                if conflicts > 0:
                    # formatted_pairs.append(f"i: {i+1}, j: {j+1}")
                    print((f"i: {i+1}, j: {j+1}"))
                else:
                        
                  xs.append(i)
                  ys.append(j)    

    print("Group conflict matrix generated successfully!")
    return group_conflicts,  xs, ys

#       
group_members = []
for i in range(0, len(random_names), 5):
    group_members.append([j for j in range(i, i + 5)])

#            
group_conflicts,xs, ys = group_conflict_matrix(group_members, conflict_matrix)

print("Group conflict matrix:")
print(group_conflicts)



```
